```python
# import
from sklearn import datasets
import pandas as pd
cancer = datasets.load_breast_cancer()

# dataframe
cancer_df = pd.DataFrame(data=cancer.data, columns = cancer.feature_names)
cancer_df['target'] = cancer.target
cancer_df.head(5)
```

```python
X = cancer_df.drop('target', axis=1)
y = cancer['target']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
# Scaling
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
std_scale.fit(X_train)
X_train_std = std_scale.transform(X_train)
X_test_std = std_scale.transform(X_test)
```

```python
# training data
# logistic regression
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression(penalty = 'l2') # penalty: Lidge
logistic.fit(X_train_std, y_train)
```

```python
# cofficients
print(logistic.coef_)

# intercept
print(logistic.intercept_)
```

```python
# predict data
pred_logistic = logistic.predict(X_test_std)
print(pred_logistic)

# class probability
pred_proba = logistic.predict_proba(X_test_std)
print(pred_proba)
```

```python
# precision_score
from sklearn.metrics import precision_score
precision = precision_score(y_test, pred_logistic)
print(precision)

# confusion matrix
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test, pred_logistic)
print(conf_matrix)

# classification_report
from sklearn.metrics import classification_report
class_report = classification_report(y_test, pred_logistic)
print(class_report)
```
