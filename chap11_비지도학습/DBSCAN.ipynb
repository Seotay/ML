```python
# load data
from sklearn.datasets import make_moons 
X, y = make_moons(n_samples=300, # features_n = 2
                  noise=0.05,
                  random_state=0)
print(X.shape)
print(y.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1],
            c='grey',
            edgecolors='black',
            marker='o')
```

```python
# DBSCAN
from sklearn.cluster import DBSCAN
dbs = DBSCAN(eps=0.2, min_samples=5)
dbs.fit(X)
label_dbs = dbs.labels_
print(label_dbs)
```

```python
# visualize data
import pandas as pd
dbs_columns = ['dbs_comp1', 'dbs_comp2']
X_dbs_df = pd.DataFrame(X, columns = dbs_columns)
X_dbs_df['target'] = y
X_dbs_df['label_dbs'] = label_dbs
X_dbs_df.head(5)
```

```python
# type of target class
print(set(X_dbs_df['target']))
print(set(X_dbs_df['label_dbs']))
```

```python
# data plot from DBSCAN
df = X_dbs_df
markers = ['o', 'x']

for i, mark in enumerate(markers):
    df_i = df[df['label_dbs'] == i]
    target_i = i
    X1 = df_i['dbs_comp1']
    X2 = df_i['dbs_comp2']
    plt.scatter(X1, X2,
                marker = mark,
                label=target_i)
plt.xlabel('dbs_component1')
plt.xlabel('dbs_component2')
plt.legend()
plt.show()
```

```python
# data plot from actual target
df = X_dbs_df
markers = ['o', 'x']

for i, mark in enumerate(markers):
    df_i = df[df['target'] == i]
    target_i = i
    X1 = df_i['dbs_comp1']
    X2 = df_i['dbs_comp2']
    plt.scatter(X1, X2,
                marker=mark,
                label=target_i)
plt.xlabel('dbs_component1')
plt.ylabel('dbs_component2')
plt.legend()
plt.show()
```

```python
# evaluate model
from sklearn.metrics import silhouette_score
sil_score = silhouette_score(X, label_dbs)
print(sil_score)
```
